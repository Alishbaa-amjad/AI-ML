## ⭐ Task 1 — Calculate Basic Statistics

**Dataset:** Daily temperatures (°C) recorded for 10 days:
```python

```

**Your job:**
1. Calculate the **mean** temperature
2. Calculate the **median** temperature

4. Print all three with clear labels

**Expected output:**

**Dataset:** Marks of 12 students:


**Your job:**
1. Find **Q1**, **Q2 (median)**, **Q3**


## ⭐ Task 4 — Detect Outliers Using IQR

**Dataset:** Annual salaries at a small company (one person is the CEO!):
```python
salaries = [30000, 32000, 35000, 33000, 200000, 31000, 34000, 36000, 29000, 32500]
```

**Your job:**
1. Compute Q1, Q3, and IQR
2. Compute the **outlier fences**:
   - Lower fence = Q1 − 1.5 × IQR
   - Upper fence = Q3 + 1.5 × IQR
3. Loop through salaries and **print any outliers**

**Expected output:**
```
Lower fence : 22,875.00
Upper fence : 43,625.00
🚨 Outlier detected: 200,000
```

> 💡 **Hints:**
> - Use `for s in salaries:` loop
> - Check: `if s < lower or s > upper:`
> - 200,000 (the CEO salary) should be the outlier
> - Use `f'{s:,}'` to format numbers with commas

````python
# Task 4
salaries = [30000, 32000, 35000, 33000, 200000,
            31000, 34000, 36000, 29000, 32500]

# Q1, Q3, IQR
q1 = np.percentile(salaries, 25)
q3 = np.percentile(salaries, 75)
# IQR
iqr = q3 - q1
print(f'idr:{iqr}')
# Outlier fences
low_fence  = q1 - 1.5 * iqr
high_fence = q3 + 1.5 * iqr
print(f'low_fence:{low_fence}')
print(f'high_fence:{high_fence}')
print(f'\nOutlier fences: {low_fence:.1f}  to  {high_fence:.1f}')
# Detect and print outliers
outliers = [x for x in salaries if x < low_fence or x > high_fence]
print(f'Outliers found: {outliers if outliers else "None"}')
````

## ⭐⭐ Task 5 — Z-Score Normalisation

**Dataset:** Heights (cm) of 10 people:
```python
heights = [150, 162, 170, 155, 180, 175, 168, 158, 172, 165]
```

**Your job:**
1. Compute the **mean** and **std** of heights
2. Calculate the **z-score** for each height
3. Print each height with its z-score
4. Flag anyone with **|z| > 1.5** as `'⚠️ Unusual'`

**Expected output:**
```
Height  Z-Score  Flag
 150     -1.84   ⚠️ Unusual
 162     -0.68
 170     +0.09
 ...
 180     +1.08
```

> 💡 **Hints:**
> - `z = (x - mean) / std`
> - Loop: `for h in heights:`
> - `flag = '⚠️ Unusual' if abs(z) > 1.5 else ''`
> - This is exactly what `StandardScaler` does in scikit-learn!

````python
# Task 5 — Your code here
heights = [150, 162, 170, 155, 180, 175, 168, 158, 172, 165]

# Compute mean and std
mean = np.mean(heights)
std  = np.std(heights)
print(f'Standard Deviation: {std:.2f}')
print(f'Mean: {mean}')
# Print table header
print(f'  {"Height":>6}  {"Z-Score":>8}  Flag')
print('-' * 30)
# Calculate and print z-scores
# for h in heights:
for h in heights:
    z = (h - mean) / std
    if z < -1.5:
        label = '← below average'
    elif z > 1.5:
        label = '→ above average'
    else:
        label = '  near average'
#     print(...)
    print(f'{h:>6}  {z:>+8.2f}  {label}')
````

## ⭐⭐ Task 6 — Explore Skewness

**Create these two datasets:**
```python
right_skewed = [2, 3, 3, 4, 4, 4, 5, 5, 10, 15, 20]
left_skewed  = [1, 5, 10, 15, 16, 16, 17, 17, 18, 18, 19]
```

**Your job:**
1. Compute **skewness** for both datasets using `stats.skew()`
2. Compute **mean** and **median** for both
3. Print results and verify the rule:
   - Right skew → mean **>** median
   - Left skew  → mean **<** median

**Expected output:**
```
--- Right Skewed ---
Skewness: +1.56   direction: positive →
Mean: 7.73   Median: 4.0
Mean > Median ✓  (right skew rule holds!)

--- Left Skewed ---
Skewness: -1.38   direction: negative ←
Mean: 13.0   Median: 16.0
Mean < Median ✓  (left skew rule holds!)
```

> 💡 **Hints:**
> - `stats.skew(data)` returns a number
> - Positive = right skew, Negative = left skew
> - Check the mean vs median relationship each time

````python

# Task 6 — Your code here
right_skewed = [2, 3, 3, 4, 4, 4, 5, 5, 10, 15, 20]
left_skewed  = [1, 5, 10, 15, 16, 16, 17, 17, 18, 18, 19]

# For right_skewed:
mean_r = np.mean(right_skewed)
print(f'Mean={mean_r:.2f}')
median_r = np.median(right_skewed)
print("Median:", median_r)
skew_r = stats.skew(right_skewed)
print("Right Skewed")
print(f'skewness={skew_r:.2f}')

if skew_r > 0:
    print("Direction: Positive →")
if mean_r > median_r:
    print("Mean > Median (Right skew rule holds!)")
else:
    print("Right skew rule not holds!")
print("\n")

# For left_skewed:
mean_l = np.mean(left_skewed)
print(f'Mean={mean_l:.2f}')
median_l = np.median(left_skewed)
print("Median:", median_l)
skew_l = stats.skew(left_skewed)
print("Left Skewed")
print(f'skewness={skew_l:.2f}')

if skew_l < 0:
    print("Direction: Negative ←")
if mean_l < median_l:
    print("Mean < Median (Left skew rule holds!)")
````

## ⭐⭐ Task 7 — Correlation Analysis

**Dataset:** Hours of sleep and test scores for 10 students:
```python
hours_sleep = [5, 6, 6, 7, 7, 8, 8, 9, 9, 10]
test_score  = [55, 60, 62, 68, 71, 75, 78, 82, 85, 88]
```

**Your job:**
1. Calculate the **Pearson correlation** and **p-value**
2. Determine strength: Strong (>0.8) / Moderate (>0.5) / Weak
3. State if the result is statistically significant (p < 0.05)
4. **Bonus:** Plot a scatter plot with a trend line

**Expected output:**
```
Pearson r = 0.997
p-value   = 0.000001
Strength : STRONG positive correlation
Significant: YES ✅
```

> 💡 **Hints:**
> - `corr, pval = stats.pearsonr(x, y)`
> - Use `if/elif/else` for strength labels
> - Bonus: `plt.scatter(x, y)` then `np.polyfit(x, y, 1)` for trend line

````python
import matplotlib.pyplot as plt
# Task 7 — Your code here
hours_sleep = [5, 6, 6, 7, 7, 8, 8, 9, 9, 10]
test_score  = [55, 60, 62, 68, 71, 75, 78, 82, 85, 88]

# 1. Pearson correlation
corr, pval = stats.pearsonr(hours_sleep, test_score)
print(f'Pearson r  = {corr:.4f}')
print(f'p-value    = {pval:.8f}')

# 2. Strength classification
if abs(corr) >= 0.8:
    strength = 'STRONG positive correlation'
elif abs(corr) >= 0.5:
    strength = 'MODERATE positive correlation'
else:
    strength = 'WEAK positive correlation'
print(f'Strength : {strength}')
# 3. Significance check
if pval<0.05:
    print(f'Significant:YES ✅')
else:
    print("Significant:NO ❌")

# 4. (Bonus) Scatter plot
plt.scatter(hours_sleep,test_score)
plt.xlabel("hours of sleep")
plt.ylabel("test score")
plt.title("sleep vs test score")
plt.show()
````

## ⭐⭐⭐ Task 9 — Hypothesis Testing

**Scenario:** A school tested a new teaching method.

```python
old_method = [65, 70, 68, 72, 74, 69, 71, 73, 66, 75]
new_method = [78, 82, 80, 85, 79, 83, 81, 84, 77, 86]
```

**Your job:**
1. State **H₀** and **H₁** in plain English (as print statements)
2. Run an **independent samples t-test**
3. Print t-statistic and p-value
4. Print the means of both groups
5. Write a clear **conclusion** based on the p-value

**Expected output:**
```
H₀: Both methods produce the same results
H₁: The new method is significantly different

Old method mean: 70.3
New method mean: 81.5
t-statistic    : -10.470
p-value        : 0.000000

✅ Reject H₀ — new method is significantly better!
```

> 💡 **Hints:**
> - `stats.ttest_ind(group1, group2)`
> - If p < 0.05 → reject H₀
> - Compare the means to see which group performed better
> - The t-stat being negative just means group1 < group2

````python
# Task 9 
from scipy import stats
import numpy as np

old_method = [65, 70, 68, 72, 74, 69, 71, 73, 66, 75]
new_method = [78, 82, 80, 85, 79, 83, 81, 84, 77, 86]

# 1. State hypotheses
print("H₀: Both methods produce the same results")
print("H₁: The new method is significantly different")

# 2. Run t-test
t_stat, p_value = stats.ttest_ind(old_method, new_method)

# 3. Print t-stat and p-value
print(f"\nt-statistic    : {t_stat:.3f}")
print(f"p-value        : {p_value:.6f}")

# 4. Print group means
old_mean = np.mean(old_method)
new_mean = np.mean(new_method)

print(f"\nOld method mean: {old_mean}")
print(f"New method mean: {new_mean}")

# 5. Conclusion
if p_value < 0.05:
    print("\nReject H₀ — new method is significantly better!")
else:
    print("\nFail to reject H₀ — no significant difference.")
````

## ⭐⭐⭐ Task 10 — MINI PROJECT: Build a Stats Report Function 🏆

**Write a function `stats_report(data, name)` that prints a complete statistical summary.**

It should output:
1. Dataset name and values
2. Mean, Median, Mode
3. Std Dev and Variance
4. Q1, Q2, Q3 and IQR
5. Skewness with direction label
6. Any outliers (using IQR method)
7. Z-scores — flag any value where |z| > 2

**Test it with:**
```python

```

**Expected output:**
```
========================================

--- Central Tendency ---
```

> 💡 **Hints:**
> - Use `def stats_report(data, name):`
> - Call all the numpy/scipy functions inside the function
> - Use `stats.mode(data).mode` and `.count` for mode
> - Build it step by step — get one section working, then add the next

> 🚀 **Bonus Challenges:**
> - Make the function **return a dictionary** instead of just printing
> - Add a **histogram** chart of the data
> - Accept a **pandas Series** as input
> - Test it on the `scores` and `salaries` datasets from earlier tasks

> 🎯 If you can build this function from scratch, you truly understand statistics for ML!

````python
# Task 10 — Your code here
# Build your stats_report function below
import numpy as np
from scipy import stats
def stats_report(data, name):
    data=np.array(data)
    print('=' * 40)
    print(f'STATS REPORT: {name}')
    print('=' * 40)
    
    # 1. Values
    print(f"Values    : {list(data)}\n")
    
    # 2. Central Tendency (mean, median, mode)
    mean = np.mean(data)
    median = np.median(data)
    mode_result = stats.mode(data, keepdims=True)
    mode = mode_result.mode[0]
    mode_count = mode_result.count[0]
    
    print("--- Central Tendency ---")
    print(f"Mean      : {round(mean,2)}")
    print(f"Median    : {median}")
    print(f"Mode      : {mode} (appeared {mode_count}x)\n")
    # 3. Spread (std, variance)
    std = np.std(data)
    var = np.var(data)
    
    print("--- Spread ---")
    print(f"Std Dev   : {round(std,2)}")
    print(f"Variance  : {round(var,2)}\n")
    # 4. Quartiles (Q1, Q2, Q3, IQR)
    q1 = np.percentile(data, 25)
    q2 = np.percentile(data, 50)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    
    print("--- Quartiles ---")
    print(f"Q1={round(q1,2)}  Q2={round(q2,2)}  Q3={round(q3,2)}  IQR={round(iqr,2)}\n")
    
    # 5. Skewness
    skew = stats.skew(data)
    
    if skew > 0:
        skew_label = "right skewed"
    elif skew < 0:
        skew_label = "left skewed"
    else:
        skew_label = "symmetric"
    
    print("--- Shape ---")
    print(f"Skewness  : {round(skew,2)}  ({skew_label})\n")
    
    # 6. Outliers
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    
    print("--- Outliers (IQR method) ---")
    outliers = data[(data < lower) | (data > upper)]
    
    if len(outliers) == 0:
        print("No outliers")
    else:
        for val in outliers:
            print(f"Outlier: {val}")
    print()
    
    # 7. Z-scores
    z_scores = stats.zscore(data)
    
    print("--- Z-Scores ---")
    for val, z in zip(data, z_scores):
        if abs(z) > 2:
            print(f"{val:>4}   z={z:+.2f}  ⚠️ EXTREME")
        else:
            print(f"{val:>4}   z={z:+.2f}")
    
    print('=' * 40)


# Test it!
stats_report([12, 15, 14, 10, 100, 13, 14, 15, 16, 11], 'Sample Data')
````
