## 🚨 Mini Project: Fraud Detection using Poisson Anomalies

A payment system records number of transactions per minute.

**Q1:** Simulate normal transaction behavior (λ = 5).

**Q2:** Inject anomalies (fraud spikes).

**Q3:** Detect anomalies using Poisson probability threshold.

**Q4 (Thinking):** How would you reduce false positives?


```python
from scipy.stats import poisson

# Q1: Simulate normal transaction behavior (lambda = 5)
np.random.seed(42)
normal_transactions = np.random.poisson(lam=5, size=100) #100 mint ka data hai har mint mei avg 5 transaction hoti h
```

```python
# Q2: Inject anomalies (fraud spikes)
# Hum kuch minutes mein achanak bohat zyada transactions dal dete hain
normal_transactions[10] = 15  # Fraud 1
normal_transactions[50] = 18  # Fraud 2
normal_transactions[85] = 20  # Fraud 3
```

```python
# Q3: Detect anomalies using Poisson probability threshold
# Hum kehte hain ke agar kisi event ki probability 0.001 se kam hai, toh wo fraud hai
threshold = 0.001
anomalies = []

for i, count in enumerate(normal_transactions):
    prob = poisson.pmf(count, mu=5)
    if prob < threshold:
        anomalies.append((i, count))

# Visualization
plt.figure(figsize=(12, 4))
plt.plot(normal_transactions, marker='o', linestyle='-', label='Transactions')
for idx, val in anomalies:
    plt.scatter(idx, val, color='red', s=100, label='Fraud Detected' if idx == anomalies[0][0] else "")
plt.axhline(y=12, color='orange', linestyle='--', label='Danger Zone')
plt.title("Fraud Detection System (Poisson Anomalies)")
plt.legend()
plt.show()

print(f"Anomalies detected at minutes: {[a[0] for a in anomalies]}")
```
