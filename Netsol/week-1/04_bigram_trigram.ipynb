Write a program to show bigram from sample code.

```python
import numpy as np
from collections import Counter
text="I want to eat I want to sleep I want to go I like to eat"
words= text.lower().split()

#bigram banana hai yaha
bigram=[(words[i],words[i+1]) 
        for i in range(len(words)-1)]

#har word ka count nikalna h yaha
bigram_counts=Counter(bigram)
word_counts=Counter(words)

# yaha predict krna h ky next word kya ho ga
def next_word(context):
    candidate=[]       #store_value
    probability=[]     #store_value_probability
    for (w1,w2), count in bigram_counts.items():
        if w1==context:
            prob=count/word_counts[w1]     #P(w2|w1)=count(w1,w2)/count(w1)
                                           #P(eat|to)=count(to,eat)/count(to)
                                           #P(to|want)=2/4 --> 1/2 --> 0.5(probability)
            candidate.append(w2)
            probability.append(prob)
    return list(zip(candidate,probability))  #zip 2 list ko pair bana deta hai
result=next_word("to")
print(result)
```

Write a program to show trigram from sample code.

```python
import numpy as np
from collections import Counter
text="I want to eat I want to sleep I want to go I like to eat"
words= text.lower().split()

#trigram banana hai yaha
trigram=[(words[i],words[i+1],words[i+2]) 
         for i in range(len(words)-2)]

#har word ka count nikalna h yaha
trigram_counts=Counter(trigram)
bigram_counts=Counter([(w1,w2) for w1,w2,w3 in trigram]) 

# yaha predict krna h ky next word kya ho ga
def next_word(a,b):
    candidate=[]       #store_value
    probability=[]     #store_value_probability
    for (w1,w2,w3), count in trigram_counts.items():
        if w1==a and w2==b:
            prob=count/bigram_counts[w1,w2]     #P(w3|w1,w2)=count(w1,w2,w3)/count(w1,w2)
                                           #P(eat|want,to)=count(want,to,eat)/count(want,to)
                                           #P(eat|want,to)=1/3 --> 0.33333(probability)
            candidate.append(w3)
            probability.append(prob)
    return list(zip(candidate,probability))  #zip 2 list ko pair bana deta hai
result=next_word("want","to")
print(result)
```

Write a program to identify the message is spam or ham. 

```python
import numpy as np
import re

emails = [("Claim your free $1M", "spam"),
    ("You won an iPhone!", "spam"),
    ("Meeting at 3 PM", "ham"),
    ("Project update attached", "ham")]

def preprocess(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", "", text)  # remove special characters
    tokens = text.split()
    return tokens

vocab = []
for email, _ in emails:
    tokens = preprocess(email)
    for word in tokens:
        if word not in vocab:
            vocab.append(word)

# Count words in spam and ham using numpy
spam_counts = np.zeros(len(vocab))
ham_counts = np.zeros(len(vocab))

for email, label in emails:
    tokens = preprocess(email)
    for word in tokens:
        idx = vocab.index(word)
        if label == "spam":
            spam_counts[idx] += 1
        else:
            ham_counts[idx] += 1

# Prior probabilities
P_spam = sum([1 for _, label in emails if label=="spam"]) / len(emails)
P_ham = sum([1 for _, label in emails if label=="ham"]) / len(emails)

# Function to classify email
def classify_email(email):
    tokens = preprocess(email)
    spam_prob = np.log(P_spam)
    ham_prob = np.log(P_ham)
    
    for word in tokens:
        if word in vocab:
            idx = vocab.index(word)
            # Laplace smoothing
            spam_prob += np.log(spam_counts[idx] + 1)
            ham_prob += np.log(ham_counts[idx] + 1)
        else:
            # unknown words
            spam_prob += np.log(1)
            ham_prob += np.log(1)
    
    if spam_prob > ham_prob:
        return "spam"
    else:
        return "ham"

# Interactive check
user_email = input("Type your message: ")
result = classify_email(user_email)
print(f"\nResult: Your email is classified as -> {result.upper()}")
```
